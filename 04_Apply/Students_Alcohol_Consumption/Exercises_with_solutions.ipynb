### 步骤8. 你注意到原始数据框仍为小写了吗？为什么会这样？修复它并将Mjob和Fjob的首字母大写。

（说明：翻译时做了以下处理：
1. 保持Markdown标题格式不变
2. 将英文问句转换为符合中文表达习惯的疑问句

### 步骤9. 创建一个名为majority的函数，该函数会返回一个布尔值到名为legal_drinker的新列中（将majority定义为年龄大于17岁）

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head()
```
